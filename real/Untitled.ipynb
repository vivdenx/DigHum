```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
defendant_csv = '../data/complete_defendants.csv'
```

```python
defendant_df = pd.read_csv(defendant_csv)
```

```python
number_list = list()
year_list = list()

for year, df in defendant_df.groupby('year'):
    number_list.append(len(df))
    year_list.append(year)
    
print(year_list)
```

```python
plt.plot(year_list, number_list)
plt.ylabel('Number of entries')
plt.xlabel('Year')
```
