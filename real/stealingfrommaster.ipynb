```python
path = '../data/stealingfrommaster_gender.csv'
import pandas as pd
df =pd.read_csv(path)
df.columns = ['year', 'x', 'x', 'x', 'x', 'male', 'female']
df.drop([0,1], axis=0, inplace=True)
df.drop('x', axis=1, inplace=True)
df.year = pd.to_numeric(df.year)


plt.plot(df.year, df.male, label = 'Male', color = 'lightblue')
plt.plot(df.year, df.female, label = 'Female', color = 'pink')
plt.xlabel('Year')
plt.ylabel('Number of defendants')
plt.tight_layout()
plt.legend()
plt.savefig('../images/stealingfrommaster_gender.jpg')
plt.show()
```
